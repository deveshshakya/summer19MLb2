```python
msg = "This message was possibly given to Father Jozo. He has described never seeing a mother so sad as the Mother of God when She talked about the Bible. "
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
```

```python
# tokenizing by period
sent = sent_tokenize(msg)
```

```python
sent
```

```python
# tokenizing by words
words = word_tokenize(msg)
```

```python
len(words)
```

```python
from nltk.corpus import stopwords
```

```python
# removing the stopwords
```

```python
new_words = [i for i in words if i.lower() in stopwords.words('english')]
```

```python
len(new_words)
```

```python
import nltk
import matplotlib.pyplot as plt
```

```python
nlp = nltk.FreqDist(words)
nlp.plot()
```

```python
nlp = nltk.FreqDist(new_words)
nlp.plot()
```
