```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Student.csv')
```

```python
plt.pie(df['Marks'], labels=df['Name'])
plt.legend(title='Names', bbox_to_anchor=(1, 1))
plt.title('Marks')
plt.show()
```

```python
plt.pie(df['Study_Hours'], labels=df['Name'])
plt.legend(title='Names', bbox_to_anchor=(1, 1))
plt.title('Study Hours')
plt.show()
```
